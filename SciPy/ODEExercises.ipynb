### Question 1

(a) Use the numerical solution to see what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs. t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 or π ?

```python
from scipy.integrate import odeint
```

```python
def deriv(theta_current,t):  # return derivatives of the array theta
    g = 9.8 #m/s^2
    ell = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])
```

```python
t = np.arange(0.,20.+0.01,0.01)

omega0 = 0. #starting from rest, omega = 0
th0 = np.linspace(0,np.pi,5)
theta_init = [0,0,0,0,0]
theta = [0,0,0,0,0]

for i in range(5):
    theta_init[i] = np.array([th0[i], omega0]) # initial value
    theta[i] = odeint(deriv,theta_init[i],t)

#plt.plot(t,theta_app(t),'b-',label="Approximate solution")
plt.plot(t,theta[0][:,0],'r',label=r'$\theta_0 =0$')
plt.plot(t,theta[1][:,0],'b',label=r'$\theta_0 =\frac{\pi}{4}$')
plt.plot(t,theta[2][:,0],'g',label=r'$\theta_0 =\frac{\pi}{2}$')
plt.plot(t,theta[3][:,0],'orange',label=r'$\theta_0 =\frac{3\pi}{4}$')
plt.plot(t,theta[4][:,0],'k',label=r'$\theta_0 =\pi$')

plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.ylim(-4.,4)
plt.xlim(0,20)
plt.grid()
plt.legend()
plt.show()
```

As the initial angle increases from $\frac{\pi}{4}$ to $\frac{3\pi} {4}$ the period gets bigger. When the initial angle is zero the pendulum is in stable equilibrium at zero. When the initial angle is at $\pi$ it is in unstable equilibrium. 

(b) Make plots of θ vs. t, ω vs. t, and ω vs. θ (the so-called "phase plot") for θ0 = 0.95π. Look carefully at their behavior. What is going on in these figures?

```python


t = np.arange(0.,10.+0.01,0.01)

omega0 = 0. #starting from rest, omega = 0
th0 = .95*np.pi

theta_init = np.array([th0, omega0]) # initial values
theta = odeint(deriv,theta_init,t)

plt.subplot(2,1,1)
plt.plot(t,theta[:,0])
plt.ylabel(r'$\theta (rad)$',fontsize=15)
plt.ylim(-4,4)

plt.subplot(2,1,2)
plt.plot(t,theta[:,1])
plt.ylabel(r'$\omega (rad/s)$',fontsize=15)
plt.xlabel('$time (s)$',fontsize = 15)


```

```python
plt.plot(theta[:,0],theta[:,1])
plt.ylabel(r'$\omega (rad/s)$',fontsize=15)
plt.xlabel(r'$\theta (rad)$',fontsize=15)
plt.ylim(-10,10)
plt.xlim(-4,4)
plt.show()
```

The peaks in the angular velocity correspond to the pendulum passing through theta = 0. Since the pendulum is oscillating without damping the phase space trajectory is a closed orbit. 

(c) For the θ0 = 0.95π case, check that energy is conserved by plotting E/m vs. t (where E/m=gℓ(1−cos(θ))+12ℓ2ω2 ).
Is the energy a constant? If not, then the computational errors are producing a non-physical loss (or gain) of energy. Review the documentation for odeint and change the error tolerance value to a tighter number.

```python
g = 9.8 #m/s^2
ell = 0.5 #m
E = g*ell*(1-np.cos(theta[:,0])) + .5*(ell**2)*(theta[:,1]**2)
plt.plot(t,E)
plt.ylim(8,12)
```

(d) Now look at what happens if we start the pendulum straight up, θ=π. Go for at least 50 s. Be sure to describe what happens numerically according to the θ vs. t, ω vs. t and E/m vs. t plots. What do you think should happen? What happens if you change the error tolerance on odeint? How does this compare to a nearly vertical pendulum: θ0=π−0.000001 ?

```python
t = np.arange(0.,50.+0.01,0.01)

omega0 = 0. #starting from rest, omega = 0
th0 = np.pi

theta_init = np.array([th0, omega0]) # initial values
theta = odeint(deriv,theta_init,t,rtol = 1.49012e-8)

E = g*ell*(1-np.cos(theta[:,0])) + .5*(ell**2)*(theta[:,1]**2)


plt.plot(t,theta[:,0],label = 'theta')
plt.plot(t,theta[:,1],label = 'omega')
plt.plot(t,E,label = 'E/m')
plt.legend(loc = 'best')
plt.ylim(-40,12)
plt.show()
```

```python
t = np.arange(0.,50.+0.01,0.01)

omega0 = 0. #starting from rest, omega = 0
th0 = np.pi

theta_init = np.array([th0, omega0]) # initial values
theta = odeint(deriv,theta_init,t,rtol = 1.49012e-10)

E = g*ell*(1-np.cos(theta[:,0])) + .5*(ell**2)*(theta[:,1]**2)


plt.plot(t,theta[:,0],label = 'theta')
plt.plot(t,theta[:,1],label = 'omega')
plt.plot(t,E,label = 'E/m')
plt.legend(loc = 'best')
plt.ylim(-40,12)
plt.xlabel(r'$t (sec)$',fontsize = 15)
plt.show()
```

```python
t = np.arange(0.,50.+0.01,0.01)

omega0 = 0. #starting from rest, omega = 0
th0 = np.pi-0.000001

theta_init = np.array([th0, omega0]) # initial values
theta = odeint(deriv,theta_init,t,rtol = 1.49012e-8)

E = g*ell*(1-np.cos(theta[:,0])) + .5*(ell**2)*(theta[:,1]**2)


plt.plot(t,theta[:,0],label = 'theta')
plt.plot(t,theta[:,1],label = 'omega')
plt.plot(t,E,label = 'E/m')
plt.legend(loc = 'best')
plt.ylim(-40,12)
plt.show()
```

From the first plot it is clear that when the pendulum is vertical it is in unstable equilibrium. It remains vertical for some time and then falls, making a full revolution and returning to its original position. The motion continues this cycle. When the tolerance is decreased, the direction that the pendulum falls alternates on each revolution. When the initial angle is slightly different than 180 degrees, the motion is similar to the motion with less tolerance. 

### Question 2

(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.
dydx=x+2y
Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.


```python
xmax = 10
dx = .25
nsteps = int(xmax/dx)
x = np.zeros(nsteps)
y = np.zeros(nsteps)

x[0]=0.
y[0]=0.

for i in range(len(x)-1):
    x[i+1] = x[i] + dx
    y[i+1] = y[i] + dx*(x[i]+2*y[i])


print "y(1) = " + str(y[x==1][0])+" (using .25 step size)"
```

```python
xmax = 10
dx = .02
nsteps = int(xmax/dx)
x2 = np.zeros(nsteps)
y2 = np.zeros(nsteps)

x2[0]=0.
y2[0]=0.

for i in range(len(x2)-1):
    x2[i+1] = x2[i] + dx
    y2[i+1] = y2[i] + dx*(x2[i]+2*y2[i])

mask = (x2 <1.01) & (x2>.99)
print "y(1) = " + str(y2[mask][0])+" (using .02 step size)"
```

(b) Plot the results from part (a) for the two different step sizes with the exact solution
y=0.25e2x−0.5x−0.25
on one plot. Include a legend. Which step size gives a more accurate result?

```python

y_exact = .25*np.exp(2*x2)-.5*x2-.25
plt.plot(x,y,label='.25 step size')
plt.plot(x2,y2,label='.02 step size')
plt.plot(x2,y_exact,label='exact')
plt.legend(loc='best')
plt.ylabel(r'$y$',fontsize=20)
plt.xlabel(r'$x$',fontsize=20)
```

The smaller step size gives a more accurate result. 

(c) Now solve the same problem using Euler-Cromer and the Midpoint method for the same pair of step sizes and plot the results along with the exact solution. Then also plot the difference between the approximate numerical solutions and the exact solution. Which combination of method and step-size gives the best result?

#### Euler-Cromer Method

```python
xmax = 10
dx = .25
nsteps = int(xmax/dx)
x = np.zeros(nsteps)
y = np.zeros(nsteps)

x[0]=0.
y[0]=0.

for i in range(len(x)-1):
    x[i+1] = x[i] + dx
    y[i+1] = y[i] + dx*(x[i+1]+2*y[i+1])


print "y(1) = " + str(y[x==1][0])+" (using .25 step size)"
```

```python
xmax = 10
dx = .02
nsteps = int(xmax/dx)
x3 = np.zeros(nsteps)
y3 = np.zeros(nsteps)

x3[0]=0.
y3[0]=0.

for i in range(len(x3)-1):
    x3[i+1] = x3[i] + dx
    y3[i+1] = y3[i] + dx*(x3[i+1]+2*y3[i+1])

mask = (x3 <1.01) & (x3>.99)
print "y(1) = " + str(y3[mask][0])+" (using .02 step size)"
```

```python
y_exact = .25*np.exp(2*x2)-.5*x2-.25
plt.plot(x,y,label='.25 step size')
plt.plot(x3,y3,label='.02 step size')
plt.plot(x2,y_exact,label='exact')
plt.legend(loc='best')
plt.ylabel(r'$y$',fontsize=20)
plt.xlabel(r'$x$',fontsize=20)
```

#### Midpoint Method

```python
xmax = 10
dx = .25
nsteps = int(xmax/dx)
x = np.zeros(nsteps)
y = np.zeros(nsteps)

x[0]=0.
y[0]=0.

for i in range(len(x)-1):
    x[i+1] = x[i] + dx
    y[i+1] = y[i] + dx*.5*((x[i+1]+2*y[i+1])+(x[i]+2*y[i]))


print "y(1) = " + str(y[x==1][0])+" (using .25 step size)"
```

```python
xmax = 10
dx = .02
nsteps = int(xmax/dx)
x4 = np.zeros(nsteps)
y4 = np.zeros(nsteps)

x4[0]=0.
y4[0]=0.

for i in range(len(x4)-1):
    x4[i+1] = x4[i] + dx
    y4[i+1] = y4[i] + dx*.5*((x4[i]+2*y4[i])+(x4[i+1]+2*y4[i+1]))

mask = (x4 <1.01) & (x4>.99)
print "y(1) = " + str(y4[mask][0])+" (using .02 step size)"
```

```python
y_exact = .25*np.exp(2*x2)-.5*x2-.25
plt.plot(x,y,label='.25 step size')
plt.plot(x4,y4,label='.02 step size')
plt.plot(x2,y_exact,label='exact')
plt.legend(loc='best')
plt.ylabel(r'$y$',fontsize=20)
plt.xlabel(r'$x$',fontsize=20)
```

```python
diff1 = abs(y_exact - y2)
diff2 = abs(y_exact - y3)
diff3 = abs(y_exact - y4)
plt.plot(x2,diff1,label='Euler Method')
plt.plot(x2,diff2,label='Euler-Cromer Method')
plt.plot(x2,diff3,label='Midpoint Method')
plt.legend(loc='best')
plt.ylabel(r'difference from exact',fontsize=20)
plt.xlabel(r'$x$',fontsize=20)
```

The Euler method with a step size of .02 gave the best result.
