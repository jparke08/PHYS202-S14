### Question 1

(a) Use the numerical solution to see what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs. t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 or π ?

```python
from scipy.integrate import odeint
```

```python
def deriv(theta_current,t):  # return derivatives of the array theta
    g = 9.8 #m/s^2
    ell = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])
```

```python
t = np.arange(0.,20.+0.01,0.01)

omega0 = 0. #starting from rest, omega = 0
th0 = np.linspace(0,np.pi,5)
theta_init = [0,0,0,0,0]
theta = [0,0,0,0,0]

for i in range(5):
    theta_init[i] = np.array([th0[i], omega0]) # initial value
    theta[i] = odeint(deriv,theta_init[i],t)

#plt.plot(t,theta_app(t),'b-',label="Approximate solution")
plt.plot(t,theta[0][:,0],'r',label=r'$\theta_0 =0$')
plt.plot(t,theta[1][:,0],'b',label=r'$\theta_0 =\frac{\pi}{4}$')
plt.plot(t,theta[2][:,0],'g',label=r'$\theta_0 =\frac{\pi}{2}$')
plt.plot(t,theta[3][:,0],'orange',label=r'$\theta_0 =\frac{3\pi}{4}$')
plt.plot(t,theta[4][:,0],'k',label=r'$\theta_0 =\pi$')

plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.ylim(-4.,4)
plt.xlim(0,20)
plt.grid()
plt.legend()
plt.show()
```

As the initial angle increases from $\frac{\pi}{4}$ to $\frac{3\pi} {4}$ the period gets bigger. When the initial angle is zero the pendulum is in stable equilibrium at zero. When the initial angle is at $\pi$ it is in unstable equilibrium.

### Question 2

(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.
dydx=x+2y
Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.


```python
xmax = 10
dx = .25
nsteps = int(xmax/dx)
x = np.zeros(nsteps)
y = np.zeros(nsteps)

x[0]=0.
y[0]=0.

for i in range(len(x)-1):
    x[i+1] = x[i] + dx
    y[i+1] = y[i] + dx*(x[i]+2*y[i])


print "y(1) = " + str(y[x==1][0])+" (using .25 step size)"
```

```python
xmax = 10
dx = .02
nsteps = int(xmax/dx)
x2 = np.zeros(nsteps)
y2 = np.zeros(nsteps)

x2[0]=0.
y2[0]=0.

for i in range(len(x2)-1):
    x2[i+1] = x2[i] + dx
    y2[i+1] = y2[i] + dx*(x2[i]+2*y2[i])

mask = (x2 <1.01) & (x2>.99)
print "y(1) = " + str(y2[mask][0])+" (using .02 step size)"
```

(b) Plot the results from part (a) for the two different step sizes with the exact solution
y=0.25e2x−0.5x−0.25
on one plot. Include a legend. Which step size gives a more accurate result?

```python

y_exact = .25*np.exp(2*x2)-.5*x2-.25
plt.plot(x,y,label='.25 step size')
plt.plot(x2,y2,label='.02 step size')
plt.plot(x2,y_exact,label='exact')
plt.legend(loc='best')
plt.ylabel(r'$y$',fontsize=20)
plt.xlabel(r'$x$',fontsize=20)
```

The smaller step size gives a more accurate result.

#### Euler-Cromer Method

```python
xmax = 10
dx = .25
nsteps = int(xmax/dx)
x = np.zeros(nsteps)
y = np.zeros(nsteps)

x[0]=0.
y[0]=0.

for i in range(len(x)-1):
    x[i+1] = x[i] + dx
    y[i+1] = y[i] + dx*(x[i+1]+2*y[i])


print "y(1) = " + str(y[x==1][0])+" (using .25 step size)"
```

```python
xmax = 10
dx = .02
nsteps = int(xmax/dx)
x3 = np.zeros(nsteps)
y3 = np.zeros(nsteps)

x3[0]=0.
y3[0]=0.

for i in range(len(x3)-1):
    x3[i+1] = x3[i] + dx
    y3[i+1] = y3[i] + dx*(x3[i+1]+2*y3[i])

mask = (x3 <1.01) & (x3>.99)
print "y(1) = " + str(y3[mask][0])+" (using .02 step size)"
```

#### Midpoint Method

```python
xmax = 10
dx = .25
nsteps = int(xmax/dx)
x = np.zeros(nsteps)
y = np.zeros(nsteps)

x[0]=0.
y[0]=0.

for i in range(len(x)-1):
    x[i+1] = x[i] + dx
    y[i+1] = y[i] + dx*.5*((x[i+1]+2*y[i])+(x[i]+2*y[i]))


print "y(1) = " + str(y[x==1][0])+" (using .25 step size)"
```

```python
xmax = 10
dx = .02
nsteps = int(xmax/dx)
x4 = np.zeros(nsteps)
y4 = np.zeros(nsteps)

x4[0]=0.
y4[0]=0.

for i in range(len(x4)-1):
    x4[i+1] = x4[i] + dx
    y4[i+1] = y4[i] + dx*.5*((x4[i]+2*y4[i])+(x4[i+1]+2*y4[i]))

mask = (x4 <1.01) & (x4>.99)
print "y(1) = " + str(y4[mask][0])+" (using .02 step size)"
```

```python
y_exact = .25*np.exp(2*x2)-.5*x2-.25
plt.plot(x,y,label='.25 step size')
plt.plot(x4,y4,label='.02 step size')
plt.plot(x2,y_exact,label='exact')
plt.legend(loc='best')
plt.ylabel(r'$y$',fontsize=20)
plt.xlabel(r'$x$',fontsize=20)
```

```python
diff1 = abs(y_exact - y2)
diff2 = abs(y_exact - y3)
diff3 = abs(y_exact - y4)
plt.plot(x2,diff1,label='Euler Method')
plt.plot(x2,diff2,label='Euler-Cromer Method')
plt.plot(x2,diff3,label='Midpoint Method')
plt.legend(loc='best')
plt.xlim(9,10)
plt.ylabel(r'difference from exact',fontsize=20)
plt.xlabel(r'$x$',fontsize=20)
```

### Question 3

(a) Create a derivative function to solve this system of equations and use odeint to solve it for a projectile fired at an angle of 50 degrees from the horizontal at an initial speed of 30 m/s. Track the projectile's motion over the time interval of 0 to 200 seconds and let (x0,y0) = (0,0). You'll have to calculate (vx0,vy0) from the information provided. Let vter=44 m/s. Plot the trajectory of the projectile (y vs. x) to see how far it can travel in the x direction before landing.

```python
def TVderiv(v_p,t):
    
    g = 9.8 #m/s^2
    vter = 44 #m/s
    
    vx = v_p[0]
    vy = v_p[1]
    x = v_p[2]
    y = v_p[3]
    dvx = -g*(vx/(vter**2))*((vx**2)+(vy**2))**(1/2)
    dvy = -g*(1+(vy/(vter**2))*((vx**2)+(vy**2))**(1/2))
    dxdt = vx
    dydt = vy
    return np.array([dvx,dvy,dxdt,dydt])
```

```python
t = np.arange(0.,200.+.1,0.1)

#initial conditions:
vx_0 = 30*np.cos(50*(np.pi/180))
vy_0 = 30*np.sin(50*(np.pi/180))
x_0 = 0
y_0 = 0

result = odeint(TVderiv,np.array([vx_0,vy_0,x_0,y_0]),t)

x = result[:,2]
y = result[:,3]

plt.plot(x,y)
plt.xlim(0,100)
plt.ylim(0,30)
plt.ylabel(r'$y (m)$', fontsize = 15)
plt.xlabel(r'$x (m)$', fontsize = 15)
```

```python
def new_TVderiv(v_p,t,vter,g):
    #g = 9.8 #m/s^2
    
    vx = v_p[0]
    vy = v_p[1]
    x = v_p[2]
    y = v_p[3]
    dvx = -g*(vx/(vter**2))*((vx**2)+(vy**2))**(1/2)
    dvy = -g*(1+(vy/(vter**2))*((vx**2)+(vy**2))**(1/2))
    dxdt = vx
    dydt = vy
    return np.array([dvx,dvy,dxdt,dydt])

def terminalVelocity(vinit,angle,time,xinit,yinit,vter,g=9.8):
    
    t = np.arange(0.,time+.1,.1)
    
    vx_0 = vinit*np.cos(angle*(np.pi/180))
    vy_0 = vinit*np.sin(angle*(np.pi/180))
    x_0 = xinit
    y_0 = yinit
    
    x_t = x_0 + vx_0*t
    y_t = y_0 + vy_0*t -.5*9.8*(t**2)
    
    initVals = np.array([vx_0,vy_0,x_0,y_0])
    result = odeint(new_TVderiv,initVals,t,args = (g,vter))
    #x = result[:,2]
    #y = result[:,3]
    
    
    plt.plot(result[:,2],result[:,3],label='With drag')
    plt.plot(x_t,y_t,label='Without drag')
    plt.ylabel(r'$y (m)$', fontsize = 15)
    plt.xlabel(r'$x (m)$', fontsize = 15)
    plt.xlim(0,100)
    plt.ylim(0,100)
    plt.legend()
    plt.grid()
    plt.show()

    
v = interact(terminalVelocity,vinit=(0.,200.),angle=(0.,90.),time=(0.,20.),xinit=(0.,10.),yinit=(0.,10.),vter=(0.,110.))    
    
```
