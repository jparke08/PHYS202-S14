```python
from random import *
from ipythonblocks import BlockGrid
from ipythonblocks import colors
import numpy as np
from IPython.html.widgets import interact, interactive
from IPython.html import widgets 
from IPython.display import display 
from IPython.display import clear_output


############################################

def random_row(grid):
    return randint(0, grid.height - 1)

def random_col(grid):
    return randint(0, grid.width - 1)

def make_system(pop_density,blue_concentration,size):  
    """This function creates an ipython block grid with
    black blocks representing empty cells, blue blocks
    representing blue agents and red blocks representing
    red agents. The first argument is pop_density which 
    defines the percentage of cells that are empty. The
    second argument is blue_concentration which defines 
    the percentage of blue agents out of the total amount of
    agents. The third argument is size which is the amount
    of blocks on each side of the grid."""
    
    system = BlockGrid(size, size, fill=(0,0,0),block_size=300/size)

    total = size**2                                  #total= total amount of blocks  
    empties = (total * pop_density)/100
    blues_reds = total - empties                     #blues_reds= total amount of agents
    blues = (blue_concentration * blues_reds)/100
    reds = blues_reds - blues

    
    #create random locations for blue agents
    count_blues = 0
    while count_blues < blues:               #iterates through random locations of blockgrid 
        row = random_row(system)             #until the right amount of blocks are made blue
        col = random_col(system)
        if system[row,col].rgb == empty:     #only makes the block blue if it is empty
            system[row,col] = blue
            count_blues += 1
    
    #create random locations for red agents    
    count_reds = 0
    while count_reds < reds:                 #iterates through random locations of blockgrid
        row = random_row(system)             #until the right amount of blocks are red
        col = random_col(system)
        if system[row,col].rgb == empty:     #only makes the block red if it is empty
            system[row,col] = red
            count_reds += 1
    system.show()
    return system,pop_density,blue_concentration,size


def find_similar_agents(grid,y,x):
    """Checks all adjacent cells and counts all the similar
    agents. At the end the function returns the percentage
    of similar agents. If the cell is empty the function
    returns None."""
    count = 0
    checks = 0
    if grid[y,x].rgb != empty:
        if x > 0:
            if grid[y,x-1].rgb != empty:
                checks += 1
            if grid[y,x]==grid[y,x-1]: #checks agent to left
                count += 1
        if x < grid.width - 1:
            if grid[y,x+1].rgb != empty:
                checks += 1
            if grid[y,x]==grid[y,x+1]:  #checks agent to right
                count += 1
        if y > 0: 
            if grid[y-1,x].rgb != empty:
                checks += 1
            if grid[y,x]==grid[y-1,x]:   #checks agent above
                count += 1
        if y < grid.height - 1:
            if grid[y+1,x].rgb != empty:
                checks += 1
            if grid[y,x]==grid[y+1,x]:     #checks agent below
                count += 1
        if x > 0 and y > 0:
            if grid[y-1,x-1].rgb != empty:
                checks += 1
            if grid[y,x]==grid[y-1,x-1]:   #checks agent above and to left
                count += 1
        if x < grid.width - 1 and y > 0: 
            if grid[y-1,x+1].rgb != empty:
                checks += 1
            if grid [y,x]==grid[y-1,x+1]:   #checks agent above and to right
                count += 1
        if x > 0 and y < grid.height - 1:
            if grid[y+1,x-1].rgb != empty:
                checks += 1
            if grid[y,x]==grid[y+1,x-1]:      #checks agent below and to left
                count += 1
        if x < grid.width - 1 and y < grid.height -1:
            if grid[y+1,x+1].rgb != empty:
                checks += 1
            if grid[y,x]==grid[y+1,x+1]:      #checks agent below and to right
                count += 1
        if checks == 0:
            return 0
        else:
            return (count/float(checks)) * 100
    else:
        return None


def find_movers(grid,threshold):
    """Checks the satisfaction by calling functon find_similar_agents
    for each agent using a threshold which is defined to be the percentage 
    of similar agents necessary for the agent to be satisfied. If the agent 
    is not satisfied it gets added to a list called movers. If the cell is 
    empty it gets added to a list called nulls."""
    nulls = []
    movers = []
    
    for j in range(grid.width):
        for i in range(grid.height):
            if find_similar_agents(grid,i,j) != None and find_similar_agents(grid,i,j) < threshold:  
                   movers.append((i,j))
            elif find_similar_agents(grid,i,j) == None:           
                nulls.append((i,j))
    return movers, nulls



def move_agents(grid,threshold,movers,nulls):
    """Each usatisfied agent in movers gets moved to an empty cell. If it is not
    satisfied in this cell, it gets moved back and tries a new cell
    until it is satisfied. If agent can't be satisfied in any cell, 
    it stays in its original position.""" 
    for i in range(len(movers)):
        for null in nulls:
            original_color = grid[movers[0]].rgb     
            grid[null] = grid[movers[0]]             #move the mover to a null and replace null with mover
            grid[movers[0]] = empty
            if find_similar_agents(grid, null[0],null[1]) < threshold: 
                grid[null] = empty              #if mover is not satisfied move it back to where it was before
                grid[movers[0]] = original_color
                if nulls[len(nulls) - 1] == null:    # if mover has no place to go then keep it where it is
                    movers.remove(movers[0])         # must remove mover so that next mover is first item on list(movers[0])
            else:
                nulls.remove(null)          # must update list of movers and nulls after swap is made
                nulls.append(movers[0])
                movers.remove(movers[0])    
                break                       #break is so that rest of nulls don't get iterated over
           

def run_shell_model(threshold,system,round):
    """This function runs a series of rounds where find_movers is
    called and the resulting movers get moved by the function
    move_agents. This continues until find_movers returns an
    empty list of movers"""

    while find_movers(system,threshold)[0]:
        movers = find_movers(system,threshold)[0]
        nulls = find_movers(system,threshold)[1]
        np.random.shuffle(movers)                  #list of movers gets shuffled before they get moved 
        np.random.shuffle(nulls)                   #list of nulls gets shuffled before agents get moved
        move_agents(system,threshold,movers,nulls)
        round += 1
        clear_output()
        print "Round %d:"%(round)
        system.show()

def find_nearest_neighbors(agent,neighbors):
    """Takes an agent and a list of it neighbors. Finds the minimum distance 
    from the agent to all of these neighbors and returns the neighbors that
    are located this distance from the agent"""
    neighbors = np.array(neighbors)
    distances = np.sqrt(((neighbors[:][0]-agent[0])**2)+((neighbors[:][1]-agent[1])**2))
    min_distance = distances.min()
    mask = distances == min_distance
    return neighbors[mask]        
        
    
def move_agents_nearest(grid,threshold,movers,nulls):
    """Finds all possible locations where the agent will be satisfied
    and adds the locations to a list called new_locs. Then moves the 
    agent to the nearest one. If there are multiple options with the 
    same minimum distance, then the location from those options is 
    chosen randomly.""" 
    for i in range(len(movers)):
        new_locs = []
        for null in nulls:
            original_color = grid[movers[0]].rgb     
            grid[null] = grid[movers[0]]             #move the agent to a null and replace null with agent
            grid[movers[0]] = empty
            if find_similar_agents(grid, null[0],null[1]) >= threshold: #finds all nulls that agent can move to 
                new_locs.append(null)                                   #adds it to list of new_locs
                
            grid[null] = empty              #moves agent back to where it was before
            grid[movers[0]] = original_color
            
            if nulls[len(nulls) - 1] == null:    
                if new_locs == []:
                    movers.remove(agents[0])
                    break                      # if agent has no place to go then keep it where it is
                else:
                    nearest = find_nearest_neighbors(movers[0],new_locs)
                    np.random.shuffle(nearest)
                    nearest = (nearest[0][0],nearest[0][1])
                    grid[nearest] = grid[movers[0]]       #after finding nearest neighbors swap with one of them randomly
                    grid[movers[0]] = empty
                    nulls.remove(nearest)          # must update list of agents and nulls after swap is made
                    nulls.append(movers[0])
                    movers.remove(movers[0]) 
        
def run_shell_model_nearest(threshold,system,round):
    """This function runs a series of rounds where find_movers is
    called and the resulting movers get moved to their nearest neighbors by the function
    move_agents. This continues until find_movers returns an
    empty list of movers"""

    while find_movers(system,threshold)[0]:
        movers = find_movers(system,threshold)[0]
        nulls = find_movers(system,threshold)[1]
        np.random.shuffle(movers)                  #list of movers gets shuffled before they get moved 
        np.random.shuffle(nulls)                   #list of nulls gets shuffled before agents get moved
        move_agents_nearest(system,threshold,movers,nulls)
        round += 1
        clear_output()
        print "Round %d:"%(round)
        system.show()
        
def start_button(widget):
    run_shell_model_nearest(threshold.value,system.result[0],round)

def reset_button(widget):
    clear_output()
    system.result = make_system(system.result[1],system.result[2],system.result[3])   # makes a new system where arguments
                                                                                      #are equal to values of interactive sliders
    


    
##################################################    
red = (255,0,0)
blue = (0,0,255)
empty = (0, 0, 0)
round = 0



threshold = widgets.FloatSliderWidget(value=41.7,description = "Threshold")               #threshold slider
display(threshold)



system = interactive(make_system,pop_density=(0,100),blue_concentration=(0,100),size=(0,50),threshold=(0.,100.))   #interactive 
display(system)


start = widgets.ButtonWidget(description='Start')         #Start button
start.on_click(start_button)
display(start)
reset = widgets.ButtonWidget(description='Reset')         #Reset button
reset.on_click(reset_button)
display(reset)
```
