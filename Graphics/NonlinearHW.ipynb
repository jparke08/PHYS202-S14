```python
plt.figure(figsize=(10, 6), dpi=80)

#Plot of Data
frequency = np.array([2.96,2.83,2.6,2.387,2.18,1.963,1.824,1.646,1.425])
gamma = .067*((2*np.pi*frequency)**2)/9.8
ticks = np.array([12,11.5,10.5,8.5,6.5,4,3,1,.5])
phi = (ticks * 5)*(np.pi/180)
plt.scatter(gamma,phi,s=5,c = 'r',marker = 'o',label='Data');

#Theoretical Plot
gamma2 = np.linspace(1,2.5,30)
phi2 = np.arccos(1/gamma2)
plt.plot(gamma2,phi2,color = 'k')
plt.plot(gamma2,-phi2,color = 'k')
plt.plot(np.linspace(0,1,2),[0,0],color = 'k',label = 'Theoretical(stable)')
plt.plot(np.linspace(1,2.5,2),[0,0],color = 'black',linestyle = '--',label= 'Theoretical(unstable)')

#Clean Up Graph
plt.xlim(0, 2.5)

ax = plt.gca() 
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.tick_params(direction='out', top=False, right=False)
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)
plt.xlabel("Gamma",fontsize=20)
plt.ylabel("Phi",fontsize=20)

plt.legend(loc ='upper left',fontsize = 10)
```
